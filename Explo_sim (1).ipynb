```python
import numpy as np
import matplotlib.pyplot as plt
D_T = 0.2*1e-12
D_R = 0.17
v = 10.*1e-6
theta_0 = 0
x_0 = 0
y_0 = 0
plt.rcParams["figure.figsize"] = (7,7)

R2_t = []
for t in range(1,10):
    t_end = t
    N = 100
    dt = t/N

    rr = []
    for i in range(50):
        x = x_0
        y = y_0
        theta = theta_0
        xx = []
        yy = []
        for i in range(N):
                    temp_th = theta + np.sqrt(2*D_R)*np.sqrt(dt)*np.random.normal(0,1)
                    x = x + v*np.cos(theta)*dt + np.sqrt(2*D_T)*np.sqrt(dt)*np.random.normal(0,1)
                    y = y + v*np.sin(theta)*dt + np.sqrt(2*D_T)*np.sqrt(dt)*np.random.normal(0,1)
                    theta = temp_th
                    xx.append(x)
                    yy.append(y)

        plt.xlabel('$x$')
        plt.ylabel('$y$')
        plt.clf()
        plt.plot(xx, yy)
        d = x**2 + y**2
        rr.append(d)

    rr = np.array(rr)
    R2_t.append(np.mean(rr))
    
tt = np.arange(1,10)
plt.figure(2)
plt.xlabel('$t$')
plt.ylabel('$<r^2(t)>$')
plt.plot(tt, R2_t, label='numerical')

r = (2*v*v/D_R)*tt + (2*v*v/(D_R*D_R))*(np.exp(-D_R*tt) - 1)
plt.plot(tt, r, label='Analytical')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
D_T = 0.2*1e-12
D_R = 0.17
v = 10.*1e-6
w = 10

theta_0 = 0
x_0 = 0
y_0 = 0
plt.rcParams["figure.figsize"] = (7,7)

mean_x_t = []
for t in range(1,10):
    t_end = t
    N = 100
    dt = t/N
    x_r = []
    for j in range(60):
        x = x_0
        y = y_0
        theta = theta_0
        xx = []
        yy = []
        for i in range(N):
                    temp_th = theta + w*dt + np.sqrt(2*D_R)*np.sqrt(dt)*np.random.normal(0,1)
                    x = x + v*np.cos(theta)*dt + np.sqrt(2*D_T)*np.sqrt(dt)*np.random.normal(0,1)
                    y = y + v*np.sin(theta)*dt + np.sqrt(2*D_T)*np.sqrt(dt)*np.random.normal(0,1)
                    theta = temp_th
                    xx.append(x)
                    yy.append(y)

        plt.xlabel('$x$')
        plt.ylabel('$y$')
        plt.clf()
        plt.plot(xx, yy,color='red')
        x_r.append(np.abs(x))
    x_r = np.array(x_r)
    mean_x_t.append(np.mean(x_r))

mean_x_t = np.array(mean_x_t)
tt = np.arange(1,10)

plt.figure(2)
plt.xlabel('$t$')
plt.ylabel('$<x(t)>$')
plt.plot(tt, R2_t, label='numerical')

r = (v/(w**2 + D_R**2))*(D_R + np.exp(-D_R*tt)*(w*np.sin(w*tt) - D_R*np.cos(w*tt)))

plt.plot(tt, r,color='chocolate',label='Analytical')
plt.legend()
```
